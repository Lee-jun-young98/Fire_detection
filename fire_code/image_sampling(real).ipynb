# 무관씬 랜덤 샘플링

```python
path = os.getcwd()
path
```

```python
file_list = os.listdir()
```

```python
random.seed(1234)

sample_list = random.sample(file_list, 1000)
```

```python
sample_list
```

```python
for i in sample_list:
    file_source = path + "\\" + i
    file_destination = file_path + "sample\\" + i
    shutil.move(file_source, file_destination)
    print("{} 이동".format(i))
```

```python
%cd "D:\fire_datasets\Training\[라벨]무관씬"
```

```python
path = os.getcwd()
path
```

```python
for i in range(len(sample_list)):
    file_name = sample_list[i][:-4] + ".json"
    file_source = path + '\\' + file_name
    file_destination = file_path + "sample\\" + file_name
    shutil.move(file_source, file_destination)
    print("{} 이동".format(file_name))
```
