```python
import os
import random
import shutil

path = "C:\\fire_datasets\\Training"
file_1 = []
jsonfile_1 = []
count = 0
random.seed(1)
for i in random.sample(os.listdir(path + "\\1\\"), 3000):
    jsonfile_1.append(random.sample(os.listdir(path + "\\1\\"), 3000)[i][:-4] + ".json")
    file_1.append(i)
    file_source = path + "\\1\\" + i
    file_destination = "C:\\yolov5-master\\yolov5-master\\data\\images"
    shutil.move(file_source, file_destination) # 이동 실행
    count += 1
    print("{}번째 ".format(count) + i + "파일 이동완료")
    
    
```

```python
file_names = ["\\[라벨]무관씬\\", "\\[라벨]유사씬\\", "\\[라벨]화재씬\\"]
```

```python
path = "C:\\fire_datasets\\Training"
image_list = os.listdir("C:\\yolov5-master\\yolov5-master\\data\\images")
count = 0
for i in image_list:
    file = i[:-4] + ".json"
    for file_name in file_names:
        if os.path.exists(path + file_name + file):
            file_source = path + file_name + file
            file_destination = "C:\\yolov5-master\\yolov5-master\\data\\images"
            shutil.move(file_source, file_destination) # 이동 실행
            count += 1
            print("{}번째 ".format(count) + file + "파일 이동완료")
```
